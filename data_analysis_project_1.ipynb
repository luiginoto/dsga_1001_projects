## Question 9

The people who rated the two movies are not all different. We have a "within-subjects" design, i.e. repeated measures for each person. Thus, it is better to do a non-parametric test for paired samples, i.e. Wilcoxon signed-rank test. To apply this test, we have to do row-wise removal of missing data.

```python
ratings = data[['Home Alone (1990)', 'Finding Nemo (2003)']].dropna()
```

```python
ratings
```

## Question 10

```python
franchises = ['Star Wars', 'Harry Potter', 'The Matrix', 'Indiana Jones', 'Jurassic Park', 'Pirates of the Caribbean', 'Toy Story', 'Batman']
inconsistent_quality = {}
```

```python
for franchise in franchises:
    print(f'Franchise: {franchise}')
    franchise_movies = []
    franchise_ratings = []
    for movie in movies:
        if franchise in movie:
            franchise_movies.append(movie)
    ratings = data[franchise_movies].dropna()
    for movie in franchise_movies:
            franchise_ratings.append(ratings[movie])
    print(f'Movies: {franchise_movies}')
    h, p = stats.friedmanchisquare(*franchise_ratings)
    print(f'Friedman Chi Square test: test statistic = {h}, p-value = {p}')
    print()
    if p < 0.005:
        inconsistent_quality[franchise] = True
    else:
        inconsistent_quality[franchise] = False
```

```python
inconsistent_quality = {}
```

```python
for franchise in franchises:
    print(f'Franchise: {franchise}')
    franchise_movies = []
    franchise_ratings = []
    for movie in movies:
        if franchise in movie:
            franchise_movies.append(movie)
            franchise_ratings.append(data[movie].dropna())
    print(f'Movies: {franchise_movies}')
    h, p = stats.kruskal(*franchise_ratings)
    print(f'Kruskal-Wallis H-test: test statistic = {h}, p-value = {p}')
    print()
    if p < 0.005:
        inconsistent_quality[franchise] = True
    else:
        inconsistent_quality[franchise] = False
```
